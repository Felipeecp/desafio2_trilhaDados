# Análise de dados maranhenses - Tabela 1
**População residente, total e respectiva distribuição percentual, por situação do domicílio e sexo e razão de sexo, segundo os municípios e as classes de tamanho da população dos municípios - Maranhão - 2010**

```python
import pandas as pd
```

```python
df_tab1 = pd.read_excel('Dados/tab1.xls')
df_tab1.head(20)
```

# Removendo linhas sem dados relevantes e renomeando colunas

## Removendo linhas de divisão de intervalos

```python
df_tab1.shape
```

```python
df_tab1[((df_tab1.iloc[:,1].str.contains("De") == True)|(df_tab1.iloc[:,1].str.contains("Até") == True)|(df_tab1.iloc[:,1].str.contains("Mais") == True))]
```

```python
df_tab1 = df_tab1[~((df_tab1.iloc[:,1].str.contains("De") == True)|(df_tab1.iloc[:,1].str.contains("Até") == True)|(df_tab1.iloc[:,1].str.contains("Mais") == True))]
```

```python
df_tab1.shape
```

```python
df_tab1.head(20)
```

## Removendo primeiras e ultimas linhas com valores nulos

```python
df_tab1.iloc[0:7]
```

```python
df_tab1.tail()
```

```python
df_tab1.drop(df_tab1.iloc[0:7].index,inplace=True)
df_tab1.drop(df_tab1.tail(2).index, inplace=True)
df_tab1
```

## Renomeando colunas

```python
colunas = ['codigo_do_municipio','nome_municipio','total_habitantes','domicilio_urbano','domicilio_rural','homem','mulher','razao_de_sexo']
df_tab1.columns = colunas
df_tab1
```

## Verificando existencia de valores nulos

```python
print('Ocorrencia de valor nulo em cada coluna:')
df_tab1.isna().sum()
```

```python
df_tab1.dtypes
```

```python
colunas_numeric = ['total_habitantes','domicilio_urbano','domicilio_rural','homem','mulher','razao_de_sexo']
df_tab1[colunas_numeric] = df_tab1[colunas_numeric].apply(pd.to_numeric,errors='coerce')
df_tab1.dtypes
```

# Guardando novo dataframe em novo arquivo csv

```python
df_tab1.to_csv('dadosLimpos/tab_1.csv',index=False)
```

```python
total_habitantes = df_tab1['total_habitantes']
total_habitantes
```

```python
Q1 = total_habitantes.quantile(0.25)
Q3 = total_habitantes.quantile(0.75)
IIQ = Q3-Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
selecao = (total_habitantes>=limite_inferior) & (total_habitantes<=limite_superior)
df_tab1_new = df_tab1[selecao]
df_tab1_new.boxplot(['total_habitantes'])
```

```python
sns.lmplot(x='mulher',y='total_habitantes',data=df_tab1_new)
```
