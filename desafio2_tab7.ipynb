# Analise Tabela 7 - Domicilios particulares por tipo de saneamento

# Sheet 1 - Domicilios em geral

```python
import pandas as pd
```

```python
tab7 = pd.read_excel("Dados/tab7.xls")
tab7
```

## Informações do dataset antes de qualquer tratamento

```python
print(f'O dataset tem {tab7.shape[0]} linhas e {tab7.shape[1]} colunas')
print(f'O dataset tem atualmente um total de {tab7.isna().sum().sum()} dados nulos')
```

## Limpeza do dataset

**Manter apenas dados de cada cidade**

**Remover todas as linhas com dados nulos que não seja uma linha referente a alguma cidade**

```python
tab7.head(15)
```

**Observando os primeiros dados, é possivel observar que até a linha 6, temos apenas descrições dos dados da tabela.**

**A linha 7 tem dados totais do dataframe, porem será removida também inicialmente**

```python
tab7.drop(tab7.iloc[0:8].index,inplace=True)
tab7.head(15)
```

### **Removendo linhas de intervalos**

```python
selecao = tab7.iloc[:,1].str.contains("De")|tab7.iloc[:,1].str.contains("Até")|tab7.iloc[:,1].str.contains("Mais")
tab7[selecao]
```

```python
tab7 = tab7[~selecao]
tab7
```

### **Renomeando colunas e resetando indice**

```python
colunas = ['codigo_municipio','nome_municipio','total_domicilios','saneamento_adequado','saneamento_semi_adequado','saneamento_inadequado']
tab7.columns = colunas
tab7.reset_index(drop=True,inplace=True)
tab7
```

### **Trocando - por 0.0**

**As colunas referentes ao tipo de seneamento estão em porcentagem.**

**Alguns dados que seriam representados por 0%, estão representados por -, para melhor aproveitamento em graficos ou calculos estatisticos, este valor será substituido por 0.**

```python
tab7[(tab7.iloc[:,3] == "-") | (tab7.iloc[:,4] == "-") | (tab7.iloc[:,5] == "-")]
```

```python
tab7 = tab7.replace({"-":0.0})
```

```python
tab7[tab7.iloc[:,3] == 0.0]
```

## Informação do dataset após tratamentos

```python
print(f'O dataset tem {tab7.shape[0]} linhas e {tab7.shape[1]} colunas')
print(f'O dataset tem atualmente um total de {tab7.isna().sum().sum()} dados nulos')
print(f'\n Informações do dataset: ')
tab7.info()
```

# Sheet 3 - Domicilios Rurais

```python
tab7C = pd.read_excel('Dados/tab7.xls',sheet_name='tab7C')
tab7C.head(15)
```

```python
tab7C.drop(tab7C.iloc[0:8].index,inplace=True)
tab7C.drop(tab7C.iloc[224:228].index,inplace=True)
selecao = tab7C.iloc[:,1].str.contains("De")|tab7C.iloc[:,1].str.contains("Até")|tab7C.iloc[:,1].str.contains("Mais")
tab7C = tab7C[~selecao]
colunas = ['codigo_municipio','nome_municipio','total_domicilios','saneamento_adequado','saneamento_semi_adequado','saneamento_inadequado']
tab7C.columns = colunas
tab7C.reset_index(drop=True,inplace=True)
tab7C
```
